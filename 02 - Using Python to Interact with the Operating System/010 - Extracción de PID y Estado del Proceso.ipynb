# Extracción de PID y Estado del Proceso

**Escenario:** 
En un sistema de registro de eventos, un equipo de desarrolladores está trabajando en una función que extrae el PID (identificador de proceso) y el estado del proceso de las líneas de registro. Este código es esencial para analizar y monitorear el comportamiento del sistema en tiempo real.

Paso a paso del código:

1. Importación del módulo `re` para trabajar con expresiones regulares.

2. Definición de la función `extract_pid(log_line)`: Esta función toma una línea de registro como entrada.

3. Se define una expresión regular `regex` para encontrar el PID y el estado del proceso en la línea de registro. La expresión regular busca un número entre corchetes seguido de un texto.

4. Se utiliza `re.search()` para buscar la coincidencia de la expresión regular en la línea de registro.

5. Si no se encuentra ninguna coincidencia (`result is None`), la función devuelve `None`.

6. Si se encuentra una coincidencia, se formatea la salida con el PID y el estado del proceso y se devuelve como una cadena de texto.

7. Se imprimen los resultados de la función `extract_pid()` con diferentes líneas de registro utilizando la función `print`.

```python
# 1. Importación del módulo `re` para trabajar con expresiones regulares.


# 2. Definición de la función `extract_pid(log_line)`: Esta función toma una línea de registro como entrada.


    # 3. Se define una expresión regular `regex` para encontrar el PID y el estado del proceso en la línea de registro. 
    # La expresión regular busca un número entre corchetes seguido de un texto.
    # r Este prefijo antes de la cadena de la expresión regular indica que es una cadena "raw" (cruda) en Python. 
    # Esto significa que los caracteres especiales dentro de la cadena (como `\`)
    # no son tratados de manera especial, 
    # lo que es útil para expresiones regulares donde los caracteres de escape son comunes.
    # \[(\d+)\] Esta parte de la expresión regular busca un número entre corchetes `[` y `]`. 
    # \[` y `\] Representan los caracteres de corchetes literales. 
    # (\d+) Representa un grupo de captura `( ... )` que busca uno o más dígitos `\d+`. 
    # Los dígitos están representados por `\d`, y el `+` 
    # significa que el dígito debe aparecer al menos una vez. 
    # Esto permite capturar el PID (identificador de proceso) que consiste en uno o más dígitos.
    #\s*:\s* Esta parte de la expresión regular busca cero o más espacios `\s*`, 
    # seguidos de dos puntos `:` y nuevamente cero o más espacios `\s*`. 
    # \s* Representa cero o más espacios en blanco.
    # : Representa el carácter de dos puntos literal.
    # (\w+) Esta parte busca uno o más caracteres de palabra `\w+`.
    # \w Representa cualquier carácter de palabra (letras, dígitos o guiones bajos).
    # + Significa que el carácter de palabra debe aparecer al menos una vez.


    # 4. A la variable "result", le agregamos `re.search()` para buscar la coincidencia de la expresión regular en la línea de registro.
    # 5. Si no se encuentra ninguna coincidencia (`result is None`), 
       
         
             
        # 5.1 la función devuelve `None`.

    # 6. Si se encuentra una coincidencia, 
    # se formatea la salida con el PID y el estado del proceso y se devuelve como una cadena de texto.


#7. Se imprimen los resultados de la función `extract_pid()` con diferentes líneas de registro utilizando la función `print`.
print(extract_pid("July 31 07:51:48 mycomputer bad_process[12345]: ERROR Performing package upgrade")) # 12345 (ERROR)
print(extract_pid("99 elephants in a [cage]")) # None
print(extract_pid("A string that also has numbers [34567] but no uppercase message")) # None
print(extract_pid("July 31 08:08:08 mycomputer new_process[67890]: RUNNING Performing backup")) # 67890 (RUNNING)
```

```python
import re

def extract_pid(log_line):
    regex = r"\[(\d+)\]\s*:\s*(\w+)"
    result = re.search(regex, log_line)
    if result is None:
        return None
    return "{} ({})".format(result.group(1), result.group(2))

print(extract_pid("July 31 07:51:48 mycomputer bad_process[12345]: ERROR Performing package upgrade")) # 12345 (ERROR)
print(extract_pid("99 elephants in a [cage]")) # None
print(extract_pid("A string that also has numbers [34567] but no uppercase message")) # None
print(extract_pid("July 31 08:08:08 mycomputer new_process[67890]: RUNNING Performing backup")) # 67890 (RUNNING)
```
