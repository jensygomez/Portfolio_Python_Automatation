# Renombrar Archivos Eliminando Dos Puntos

**Escenario:** Supongamos que necesitas renombrar todos los archivos dentro de una carpeta y sus subcarpetas, eliminando los dos puntos de los nombres de archivo.

**Paso a paso:**
1. El programa comienza definiendo una función llamada `eliminar_dos_puntos` que toma una `ruta` como argumento.
2. Se importan los módulos `os` y `re` para manipular el sistema operativo y realizar operaciones de expresiones regulares, respectivamente.
3. Se define un patrón de expresión regular para buscar los dos puntos en los nombres de archivo.
4. Se utiliza `os.walk` para recorrer recursivamente todos los archivos y carpetas dentro de la ruta especificada.
5. Por cada archivo encontrado, se verifica si contiene dos puntos en su nombre utilizando la expresión regular definida.
6. Si el archivo contiene dos puntos, se renombra eliminando esos dos puntos del nombre.
7. Se manejan los errores si la operación de renombrado falla.
8. Se repite el proceso para las subcarpetas encontradas dentro de la ruta especificada.
9. Se solicita al usuario ingresar la ruta de la carpeta que contiene los archivos a renombrar.
10. Se llama a la función `eliminar_dos_puntos` con la ruta ingresada por el usuario para comenzar el proceso de renombrado.

```python
#!/usr/bin/env python3

import os  # Importamos el módulo os para interactuar con el sistema operativo
import re  # Importamos el módulo re para trabajar con expresiones regulares

# Definimos la funcion "eliminar_dos_puntos", que recibe como argumento "ruta"
def eliminar_dos_puntos(ruta):
    
    # Compilamos una expresión regular para buscar dos puntos en los nombres de archivo
    patron = re.compile(r':')  

    for root, dirs, files in os.walk(ruta):  
    # Iteramos sobre los archivos y directorios dentro de la ruta especificada
        for archivo in files:  
        # Iteramos sobre los archivos en la carpeta actual
            if patron.search(archivo):  
            # Si el nombre del archivo contiene dos puntos
                nuevo_nombre = patron.sub("", archivo)  
                # Eliminamos los dos puntos del nombre del archivo
                ruta_antiguo = os.path.join(root, archivo)  
                # Obtenemos la ruta completa del archivo original
                ruta_nuevo = os.path.join(root, nuevo_nombre)  
                # Obtenemos la ruta completa del nuevo archivo
                try:
                    os.rename(ruta_antiguo, ruta_nuevo)  
                    # Intentamos renombrar el archivo
                    print(f"Renombrado: {ruta_antiguo} -> {ruta_nuevo}")  
                    # Imprimimos un mensaje de éxito
                except Exception as e:
                    print(f"No se pudo renombrar {ruta_antiguo}: {e}")  
                    # Imprimimos un mensaje de error en caso de fallo

        for carpeta in dirs:  
        # Iteramos sobre los subdirectorios en la carpeta actual
            eliminar_dos_puntos(os.path.join(root, carpeta))  
            # Llamamos recursivamente a la función para procesar los subdirectorios

ruta_carpeta = input("Ingrese la ruta de la carpeta que contiene los archivos a renombrar: ")  
# Solicitamos al usuario que ingrese la ruta de la carpeta
eliminar_dos_puntos(ruta_carpeta)  
# Llamamos a la función principal para iniciar el proceso de renombrar archivos
```

```python
#!/usr/bin/env python3

import os
import re

def eliminar_dos_puntos(ruta):
    patron = re.compile(r':')

    for root, dirs, files in os.walk(ruta):
        for archivo in files:
            if patron.search(archivo):
                nuevo_nombre = patron.sub("", archivo)
                ruta_antiguo = os.path.join(root, archivo)
                ruta_nuevo = os.path.join(root, nuevo_nombre)
                try:
                    os.rename(ruta_antiguo, ruta_nuevo)
                    print(f"Renombrado: {ruta_antiguo} -> {ruta_nuevo}")
                except Exception as e:
                    print(f"No se pudo renombrar {ruta_antiguo}: {e}")

        for carpeta in dirs:
            eliminar_dos_puntos(os.path.join(root, carpeta))

ruta_carpeta = input("Ingrese la ruta de la carpeta que contiene los archivos a renombrar: ")
eliminar_dos_puntos(ruta_carpeta)
```
